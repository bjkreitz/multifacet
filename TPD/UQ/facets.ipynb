```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd

plt.rcParams['figure.figsize']=(5,6)
plt.rcParams['axes.linewidth'] = 3
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rc('axes', labelsize=20)
plt.rc('legend', fontsize=18)
plt.rcParams['lines.markersize'] = 10
plt.rcParams['xtick.direction']='in'
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.major.size']=10
plt.rcParams['xtick.major.width']=2
plt.rcParams['ytick.major.size']=10
plt.rcParams['ytick.major.width']=2
plt.rcParams['legend.edgecolor']='k'
plt.rcParams['axes.unicode_minus']=False
plt.rcParams["legend.framealpha"] = 1
plt.rcParams['xtick.major.pad'] = 15
plt.rcParams['ytick.major.pad'] = 15
plt.rcParams['legend.handletextpad']=0.4
plt.rcParams['legend.columnspacing']=0.5
plt.rcParams['legend.labelspacing']=0.3
plt.rcParams['legend.title_fontsize'] = 14
plt.rcParams['axes.formatter.limits']=(-3, 6)
plt.rcParams['hatch.linewidth'] = 5.0
plt.rcParams['axes.titlepad'] = 20 

colormap=plt.cm.Dark2
colors = [colormap(i) for i in np.linspace(0, 1, 8)]

base=np.array([0.6923,0.211,0.0439,0.0528])
diff=np.array([0.6558,0.2637,0.0694,0.0111])
nodiff=np.array([0.4025,0.4408,0.0812,0.0754])
```

```python
gs=gridspec.GridSpec(nrows=1, ncols=1)
gs.update(wspace=0.3, hspace=0.2)

ax0=plt.subplot(gs[0,0])

n=np.array([0,1,2,3])

for i in range(len(n)):
    ax0.bar(n[i]-0.3,base[i],width=0.3,color=colors[i],edgecolor='k')

    ax0.bar(n[i],diff[i],width=0.3,color=colors[i],edgecolor='w', hatch="/")
    ax0.bar(n[i],diff[i],width=0.3,color='None',edgecolor='k')
    
    ax0.bar(n[i]+0.3,nodiff[i],width=0.3,color=colors[i],edgecolor='w', hatch="o")
    ax0.bar(n[i]+0.3,nodiff[i],width=0.3,color='None',edgecolor='k')
    
ax0.set_ylabel('$\mathrm{facet\ distribution}$')
ax0.set_xticks([0,1,2,3])
ax0.set_xticklabels(["Ni(111)", "Ni(100)", "Ni(211)", "Ni(110)"], rotation=45)

ax0.bar(100, 100,width=0.3,color='k',edgecolor='w', label='$\mathrm{base\ case}$')
ax0.bar(100, 100,width=0.3,color='k',edgecolor='w', hatch="/", label='$\mathrm{best\ match\ diff}$')
ax0.bar(100, 100,width=0.3,color='k',edgecolor='w', hatch="o", label='$\mathrm{best\ match\ w/o\ diff}$')


ax0.set_xlim([-0.75,3.75])
ax0.set_ylim([0,1])

ax0.legend()

plt.savefig('facet_distribution.png', dpi=400, bbox_inches='tight', transparent=False)

plt.show()
```
